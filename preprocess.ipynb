```python
country_df = pd.read_csv('resources/world-data-2023.csv')
country_df
```

```python
country_df['GDP'] = country_df['GDP'].str.strip('$')
```

# Turn GDP and Population into numerical values

```python
country_df['GDP'] = country_df['GDP'].str.replace(',', '').astype(float)
country_df = country_df.loc[:][country_df['Population'].notna()]
country_df['Population'] = country_df['Population'].str.replace(',', '').astype(int)
```

```python
country_df
```

```python
country_df.to_csv('resources/processed/CountryData.csv', index=False)
```
